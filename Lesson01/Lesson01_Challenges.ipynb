```python
#Importing pandas library.
import pandas as pd

#Assigning the dataframe url to a variable.
url_df = "https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true"

#Assigning the uncompressed csv dataframe to a variable.
df = pd.read_csv(url_df, compression = 'zip')

#Renaming specific columns using rename(), for the purpose of translating it to English.
df.rename({'tratamento':'treatment', 'tempo':'duration', 'droga':'drug'}, axis=1, inplace=True)

#Renaming the contents of the column 'treatment', for the purpose of translating it to English.
df['treatment']= df['treatment'].replace({'com_droga':'with_drug', 'com_controle':'with_control'})

#Calling the dataframe.
df
```

### Challenge 06: Remove the hyphen (-) from the columns' names.


So, this was quite challenging.

I first found a way to iterate through the columns using for loop and  .columns, then I had to isolate the columns with the hyphen so I created an if statement and with string formatting I removed the hyphen.
To actually change the column names I used .rename(), which I used at the beginning to translate the first columns.

```python
#df.rename({'tratamento':'treatment', 'tempo':'duration', 'droga':'drug'}, axis=1, inplace=True)
for column in df.columns:
    if '-' in column:
        index = column.find('-')
        changed_column = column[:index] + column[index+1:]
        df.rename({column:changed_column}, axis=1, inplace=True)
    
       

```
